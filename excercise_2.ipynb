Generate a random sample from the functions 1 / (x + 1) and sqrt(x)

```python
import numpy as np
import matplotlib.pyplot as plt

start, end = 0, 10
xx = np.random.uniform(start, end, 50)
yy = 1 / (xx + 1)
yy2 = np.sqrt(xx)
plt.scatter(xx, yy)
```
