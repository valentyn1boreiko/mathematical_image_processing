**Task 1**

```python
from PIL import Image
import numpy as np
import glob
from utils_image import image_histogram, images_plot,\
    gradient_histogram, hess_histogram

# load the images and convert them to matrices
images = []
for file in glob.glob("Bilder-20200422/*.jpg"):
    image = Image.open(file)
    u_ij = np.asarray(image)
    item = {'title': file.split('.')[0].split('/')[1],
            'matrix': u_ij,
            'histogram': image_histogram(u_ij),
            'probability': image_histogram(u_ij, True),
            'grad_histogram': gradient_histogram(u_ij),
            'grad_probability': gradient_histogram(u_ij, True),
            'hess_histogram': hess_histogram(u_ij),
            'hess_probability': hess_histogram(u_ij, True)}
    images.append(item)
    
```

```python
# plot the images, histograms, and densities
images_plot(images)
```

We can see that adding Gaussian noise makes the histograms much smoother. 
Sampling error makes them somewhat smoother.

**Task 2**

```python
images_plot(images, 'grad_')
```

We need less pins on the x axis, as there are not many drastic contrasts 
(changes from light to dark pixels or vice versa)
in the picture.  
We can see as well that the pictures with noise produce more irregular histograms,
as they produce more irregular contrasts in the pictures.
